```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
curve_Ag = np.loadtxt("Murnagghan curve Ag.txt")
```

```python
lattice_Ag = curve_Ag[:,0]
total_E_Ag = curve_Ag[:,1]
```

```python
plt.scatter(lattice_Ag,total_E_Ag)
plt.plot(lattice_Ag,total_E_Ag)

plt.xlabel("Lattice Parameter / $\AA$")
plt.ylabel("Ag Total Energy / Ry")

plt.show()
```
