```python
import numpy as np
import matplotlib.pyplot as plt
from energydiagram import ED

plt.rcParams['figure.figsize'] = [12, 5]
```

```python
diagram = ED()

diagram.add_level(0,'H$^+$ + e$^-$')
diagram.add_level(0.012,'Ag')
diagram.add_level(0.020,'Au','last')
diagram.add_level(-0.022,'Mo','last')
diagram.add_level(-0.011,'Pt','last')
diagram.add_level(0,'1/2 H$_2$')

#diagram.plot(show_IDs=True)

diagram.add_link(0,1)
diagram.add_link(0,2)
diagram.add_link(0,3)
diagram.add_link(0,4)
diagram.add_link(1,5)
diagram.add_link(2,5)
diagram.add_link(3,5)
diagram.add_link(4,5)

diagram.plot(ylabel="Energy / Hartree") # this is the default ylabel
#diagram.plot(ylabel="Energy / $kcal$ $mol^{-1}$") # this is the default ylabel
```
