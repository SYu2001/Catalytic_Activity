```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.rcParams['figure.figsize'] = [12, 5]
plt.rcParams['font.size'] = 15
plt.rcParams['axes.labelsize'] = 15
plt.rcParams['xtick.labelsize'] = 13
plt.rcParams['ytick.labelsize'] = 13
```

```python
def del_E(n, E_surf_nH, E_surf, E_H2):
    return 1/n * (E_surf_nH-E_surf - n/2 * E_H2)
```

```python
Metal = np.array(["Ag","Au","Mo","Pt"])

E_Metal = np.array([-2369.251320050268532,-2127.526871266153194,-4365.795058538102239,-7687.143423162807267])
E_Metal_H = np.array([-2369.830342421368641,-2128.098453575051735,-4366.408591960806916,-7687.745893106200128])
n = 1
E_H2 = -1.164851668533423
```

```python
E_chemisorption = del_E(n, E_Metal_H, E_Metal, E_H2)

print(E_chemisorption)
```

```python
i_0 = np.array([-5.0,-6.6,-7.07,-3.34])

fig, ax = plt.subplots()
ax.scatter(E_chemisorption,i_0,color = 'black')

for i, txt in enumerate(Metal):
    ax.annotate(txt, (E_chemisorption[i], i_0[i]))

#plt.scatter(E_chemisorption[:2],i_0[:2])
#plt.scatter(E_chemisorption[2:],i_0[2:])

a,b = np.polyfit(E_chemisorption[:2],i_0[:2],1)
plt.plot(E_chemisorption[:2], a*E_chemisorption[:2]+b,color = 'None')

x1_ex = np.linspace(-0.014,0.013,1000)
plt.plot(x1_ex, a*x1_ex+b,color = 'black')

c,d = np.polyfit(E_chemisorption[2:],i_0[2:],1)
plt.plot(E_chemisorption[2:], c*E_chemisorption[2:]+d,color = 'None')

x2_ex = np.linspace(-0.031,-0.014,1000)
plt.plot(x2_ex,c*x2_ex+d,color = 'black')

#plt.title('Hydrogen Evolution')
plt.xlabel('$\Delta$$E_{H}$/(Hartree)')
plt.ylabel('Log($i_{0}$/(A $cm^{-2}$))')

plt.yticks(np.linspace(-8,-1,15))
plt.xticks(np.linspace(-0.035,0.015,11))

plt.show()
```
