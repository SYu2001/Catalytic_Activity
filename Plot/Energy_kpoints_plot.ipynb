```python
Energy_kpoints = np.loadtxt("Energy_kpoints.txt")
```

```python
density_MP = Energy_kpoints[:,0]
energy = Energy_kpoints[:,1]

plt.scatter(density_MP,energy)
plt.plot(density_MP,energy)

plt.xlabel('Density of MP Grid')
plt.ylabel('Energy')

plt.show()
```
